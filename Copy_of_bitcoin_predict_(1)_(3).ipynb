<h2>Adding RSI</h2>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = np.zeros_like(delta)
    avg_loss = np.zeros_like(delta)
    for i in range(1, len(delta)):
        if i < window:
            avg_gain[i] = gain[:i+1].mean()
            avg_loss[i] = loss[:i+1].mean()
        else:
            avg_gain[i] = (avg_gain[i-1] * (window - 1) + gain[i]) / window
            avg_loss[i] = (avg_loss[i-1] * (window - 1) + loss[i]) / window
    rs = avg_gain / avg_loss
    rsi = 100.0 - (100.0 / (1.0 + rs))
    rsi[:window] = np.nan
    return rsi
```

```python
# Add RSI to the dataset
data['RSI'] = calculate_rsi(data)
data['RSI'].fillna(method='bfill', inplace=True)

# Display the first few rows with RSI
print(data.head(20))
```

<h2>adding RSI</h2>

RSI values range from 0 to 100.
An RSI above 70 typically indicates that the asset is overbought.
An RSI below 30 typically indicates that the asset is oversold.
These conditions can signal potential reversal points in the price trend.

Detect Momentum:

RSI can help identify the momentum of price movements.
Increasing RSI values suggest upward momentum, while decreasing RSI values suggest downward momentum.

RSI needs 14 periods to start calculating the values, which is why the first 14 values are NaN. This is expected behavior because there is not enough data to calculate the RSI for those periods.

Improved Handling
You can handle the NaN values in several ways, depending on your specific needs:

Keep the NaN Values: This is useful if you are okay with having missing values at the beginning of your dataset.
Forward Fill the NaN Values: Fill the missing values with the first available RSI value.
Drop the NaN Values: Remove the rows with NaN values if they are not significant for your analysis.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = np.zeros_like(delta)
    avg_loss = np.zeros_like(delta)
    for i in range(1, len(delta)):
        if i < window:
            avg_gain[i] = gain[:i+1].mean()
            avg_loss[i] = loss[:i+1].mean()
        else:
            avg_gain[i] = (avg_gain[i-1] * (window - 1) + gain[i]) / window
            avg_loss[i] = (avg_loss[i-1] * (window - 1) + loss[i]) / window
    rs = avg_gain / avg_loss
    rsi = 100.0 - (100.0 / (1.0 + rs))
    rsi[:window] = np.nan
    return rsi
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Assuming 'features' and 'target' are DataFrames containing the specified columns

# Prepare the data
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
target_scaled = scaler.fit_transform(target)

# Reshape the data for LSTM input
X = np.reshape(features_scaled, (features_scaled.shape[0], 1, features_scaled.shape[1]))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, target_scaled, test_size=0.2, shuffle=False)

# Build the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Make predictions
predictions = model.predict(X_test)

# Plot actual prices and predicted prices
plt.plot(y_test, label='Actual Prices')
plt.plot(predictions, color='red', label='Predicted Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Bitcoin Price Prediction using LSTM')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Evaluate the LSTM model using accuracy metric
predictions_binary = np.where(predictions > 0.5, 1, 0)  # Convert predictions to binary values
y_test_binary = np.where(y_test > 0.5, 1, 0)  # Convert actual values to binary
accuracy = accuracy_score(y_test_binary, predictions_binary)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.metrics import r2_score

# Evaluate the LSTM model using R-squared metric
r_squared = r2_score(y_test, predictions)
print(f'R-squared: {r_squared}')
```
